### 1: Import des librairies

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:.2f}".format)
```

### 2: Chargement des données

```python
df = pd.read_csv("data/credit_card_transactions.csv")

df.head()
```

### 3: Inspection rapide

```python
df.info()
df.describe()
```

```python
df = df.drop(columns=["Unnamed: 0"])
```

```python
df["trans_date_trans_time"] = pd.to_datetime(df["trans_date_trans_time"])
df["dob"] = pd.to_datetime(df["dob"])
```

### 4: Vérification des valeurs manquantes

```python
df.isnull().sum()
```

### 5️: Vérification des doublons

```python
df.duplicated().sum()
```

### 6: Gestion de merch_zipcode

```python
df["merch_zipcode"] = df["merch_zipcode"].fillna(-1)
df["merch_zipcode"] = df["merch_zipcode"].astype(int)
```

```python
df.info()
```

### KPIs globaux

```python
total_transactions = len(df)
total_amount = df["amt"].sum()
avg_amount = df["amt"].mean()
fraud_rate = df["is_fraud"].mean() * 100

total_transactions, total_amount, avg_amount, fraud_rate
```

### Analyse par catégorie

```python
category_analysis = (
    df.groupby("category")["amt"]
    .agg(["count", "sum", "mean"])
    .sort_values("sum", ascending=False)
)

category_analysis.head()
```

### Analyse par État (contrôle géographique)

```python
state_analysis = (
    df.groupby("state")["amt"]
    .agg(["count", "sum", "mean"])
    .sort_values("sum", ascending=False)
)

state_analysis.head()
```

### Analyse FRAUDE

```python
fraud_analysis = (
    df[df["is_fraud"] == 1]
    .groupby("category")["amt"]
    .agg(["count", "sum", "mean"])
    .sort_values("count", ascending=False)
)

fraud_analysis.head()
```

### Détection d’anomalies 

##### Z-score

```python
mean_amt = df["amt"].mean()
std_amt = df["amt"].std()

df["z_score"] = (df["amt"] - mean_amt) / std_amt

anomalies = df[df["z_score"].abs() > 3]
len(anomalies)
```

#### Top Anomalies

```python
anomalies.sort_values("amt", ascending=False).head(10)
```

#### Anomalies par catégorie

```python
anomaly_by_category = (
    anomalies.groupby("category")["amt"]
    .count()
    .sort_values(ascending=False)
)

anomaly_by_category.head()
```

### Préparation pour Power BI

```python
monthly_summary = (
    df.set_index("trans_date_trans_time")
      .resample("M")
      .agg(
          total_amount=("amt", "sum"),
          avg_amount=("amt", "mean"),
          fraud_count=("is_fraud", "sum"),
          transaction_count=("amt", "count")
      )
      .reset_index()
)
```

```python
df.to_csv("data/clean_transactions.csv", index=False)
monthly_summary.to_csv("data/monthly_summary.csv", index=False)
anomalies.to_csv("data/anomalies.csv", index=False)
```
